```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
init_printing()
```

```python
x, y, a, b, r, r2 = symbols('x y a b r r2')
```

# Least squares equation

```python
ab = Matrix([a,b])
```

```python
ab     #also called beta (solution of matrix equation)
```

```python
ones_m = ones(7, 1)
```

```python
x = Matrix([113, 222, 315, 378, 523, 578, 727])
```

```python
X = ones_m.row_join(x)
```

```python
X                                                     # 'X' matrix from matrix equation
```

```python
y = Matrix([311, 459, 733, 847, 1191, 1237, 1547])        # 'y' matrix from matrix equation
```

```python
plt.scatter(x, y, s=1)
```

```python
X*ab
```

```python
r = X*ab-y     #X*ab-y
```

```python
r      #This is the residuals vector
# r = a + xi*b - yi
# a + xi*b     - predicted
# yi           - actual value
# r            - error / residual
```

```python
r.T
```

```python
r.T*r    #scalar, the sum of squares of residuals
```

```python
expand(r.T*r)    #scalar
```

```python
r2 = r.T*r
```

```python
diff(r2,a)
```

```python
diff(r2,b)
```

```python
diff(r2,a).col_join(diff(r2,b))        #a system of equations
```

```python
solve(diff(r2,a).col_join(diff(r2,b)))
```

```python
solve(r2)
```

```python
diff(r,a)   #ones
```

```python
diff(r,b)   #x
```
