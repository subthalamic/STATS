```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
normalrand = np.random.normal(0, 1, size=100)
unifrand = np.random.uniform(-1, 1, 100)
```

```python
np.std(normalrand)
```

```python
regtest = pd.DataFrame(columns=['line', 'randline'], index=range(100))
```

```python
regtest.line=range(1, 101)
```

```python
regtest['normalrand']=normalrand
regtest['unifrand']=unifrand
regtest.tail(5)
```

```python
regtest.randline=regtest.line+regtest.unifrand*10
```

```python
regtest.randline.values
```

```python
plt.scatter(regtest.line, regtest.randline, s=0.3)
plt.show()
```

```python
plt.scatter(regtest.line, regtest.line + regtest.normalrand*5, s=0.3)
plt.show()
```
