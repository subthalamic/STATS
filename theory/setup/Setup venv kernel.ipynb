# Trying to use venv installed packages in jupyter notebook

[Activate the Virtual Environment: Before launching the Jupyter notebook server, 
activate the virtual environment in your PyCharm project
where you installed the necessary Python packages. You can do this in the terminal:]

Set-ExecutionPolicy RemoteSigned
venv\Scripts\activate

[Install ipykernel: Ensure that the ipykernel package is installed in your virtual environment.
This package allows you to use the virtual environment's Python
interpreter as a kernel in Jupyter notebooks.
If it's not installed, you can install it using pip:]

pip install ipykernel

[Register Virtual Environment as Jupyter Kernel: 
Run the following command to register the virtual environment's
Python interpreter as a kernel for Jupyter:]

python -m ipykernel install --user --name=venvstats

[Verify Installation: After running the command, you should see a confirmation message
indicating that the kernel has been installed. You can also verify that the kernel
is installed by checking the list of available kernels using:]

jupyter kernelspec list

######

[Revert:]

[The deactivate command reverses the changes made by the activate command.
When you run deactivate, it restores the shell environment to its state
before the virtual environment was activated.]

deactivate
Set-ExecutionPolicy Restricted
