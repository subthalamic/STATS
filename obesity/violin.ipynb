```python
#quantiles = [0.25, 0.5, 0.75]
n_cols=7
fig, ax = plt.subplots(nrows=5, ncols=n_cols, figsize=(12, 12))
i=0
for column in data_columns:
    ax[i//n_cols,i%n_cols].violinplot(perim[column][:40], showmeans=True, showmedians=True)
    ax[i//n_cols,i%n_cols].set_title(column)
    i+=1
plt.tight_layout()
plt.savefig('violin_indiv.png')
plt.show()
```
