```python
descr = qdn.iloc[:,1:].agg(['count', 'median', 'mean', 'min', 'max', 'std', 'var'])
descr.index = ['N', 'median', 'mean', 'minimum value', 'maximum value', 'standard deviation', 'variance']
descr.columns = ['weekly high effort (days)', 'daily high effort (minutes)', 'weekly moderate effort (days)', 
                 'daily moderate effort (minutes)', 'weekly walking time (days)', 'daily walking time (minutes)', 
                'daily sitting time (minutes)', 'weekly hight effort (hours)', 'weekly moderate effort (hours)', 
                'weekly walking time (hours)', 'weekly sitting time (hours)']
descr = descr.round(2).T
descr.to_csv('descr.csv')
descr
```
